## Problématique

On cherche à identifier un type d'émetteur de communication à partir de l'observation d'un signal provenant de l'émetteur
de 2048 échantillons IQ (In Phase / Quadrature) ie le signal prend des valeurs complexes. On représente la partie
réelle et la partie imaginaire du signal par deux canaux réel d'un signal multivarié.

L'émetteur peut provenir de 6 catégories différentes.
Les paramètres différenciant les différentes catégories sont
- le type de modulation
- la présence ou non de séquences pilotes et le cas échéant la structure de trame pilotes / données
- le débit de la transmission

Les signaux se propagent en champs libre et sont enregistrés par une antenne. Le signal reçu est transposé en bande de base c'est à dire que si le signal est transmis autour d'une fréquence centrale f0, une première étape de traitement du signal à la réception recentre le signal autour de la fréquence 0.


Les différents signaux observés dans ce TP sont entachés de différentes erreurs caractéristiques de la propagation
électromagnétique comme :
- modification aléatoire de la phase du signal lors de la transmission
- imperfection de la transposition en bande de base qui laisse le signal transposé à une fréquence df0 << f0
- présence d'interférence entre les symboles transmis (dûes par exemple à plusieurs chemins de propagation)
- présence d'un bruit blanc additif gaussien

Le niveau du bruit relativement à celui du signal utile est décrit par le SNR (Signal to Noise Ratio) et exprimé en dB. On suppose que le SNR est connu lors de l'acquisition d'un signal. Lors de ce TP nous rencontrerons 4 niveaux de SNR: 30 dB (facile), 20 dB, 10 dB et 0 dB (en espérant qu'on puisse faire quelque chose de ces données).
Un de nos objectifs sera de qualifier la performance des algorithmes mis en place en fonction du SNR.

Pour entrainer des réseaux de neurones profond sur nos données nous allons utiliser le framework Pytorch.
Une première étape va consister à transférer les données de numpy à PyTorch, cela passe par deux objets :

Un dataset prend la forme
```python
class MyDataset(torch.utils.data.Dataset):
    
    def __init__(self, path_to_data):
        ...
    def __len__(self): #retourne le nombre de données dans le dataset
        ...
    def __getitem__(self,i): #retourne pour chaque indice i un couple (data_i, lablel_i), data_i étant un signal et label_i le label associé au signal
        ...
```

Implémentez une classe Dataset pour le dataset considéré ici

````python
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
class MyDataset(torch.utils.data.Dataset):

    def __init__(self, path_to_data):
        data = h5py.File(path_to_data , 'r')
        self.signals = np.array(data['signaux'])
        self.labels = np.array(data['labels'])
        data.close()

    def __len__(self): #retourne le nombre de données dans le dataset
        return self.signals.shape[0]
    def __getitem__(self,i): #retourne pour chaque indice i un couple (data_i, lablel_i), data_i étant un signal et label_i le label associé au signal
        return torch.tensor(self.signals[i]).to(device), torch.tensor(self.labels[i]).to(device)
````

Ce Modèle doit être très simple, il doit être rapide à exécuter, il servira à tester et éventuellement débugger la pipeline d'entrainement que vous mettrez en place un peu plus loin. Un template d'implémentation d'une classe Model se trouve dans les diapositives du cours.

**Réseau dumb :** Nous avons choisi comme réseau simple un réseau contenant une couche de convolution 1D avec 4 filtres, une taille de kernel de 5 et un stride de 4. Ensuite nous utilisons une couche max pooling de taille 3 pour réduire la dimension, puis nous applatissons notre vecteur pour appliquer une couche linéaire suivie d'une fonction d'activation softmax.

Le réseau est ainsi très simpliste, nous ne nous attendons donc pas à obtenir avec celui-ci des performances élevées.

````python
class DumbNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = torch.nn.Conv1d(in_channels = 2, out_channels = 4, kernel_size = 5, stride = 4)
        self.fc1 = torch.nn.Linear(680, 6)
        self.softmax = torch.nn.Softmax()
        self.pool = torch.nn.MaxPool1d(kernel_size=3)  
    def forward(self, x):
        x = x.transpose(1, 2)
        x = self.conv1(x)
        x = self.pool(x)
        x = x.view(x.size(0), -1)
        x = self.fc1(x)
        return self.softmax(x)


````

## Mise en place de la pipeline d'entraînement


Mettez en place votre pipeline et testez là sur votre modèle dumb. Faites en sorte que votre façon de logger les loss et les métriques vous permette de visualiser l'évolution de ces différents indicateurs sur l'ensemble d'entrainement et de validation au fil des epochs.

````python
def train_valid_pipeline(model_class, save_name, loss = torch.nn.NLLLoss(), n_epochs = 100, lr = 1e-6):
    model = model_class()  
        
    optimizer = torch.optim.Adam(model.parameters(), lr=lr) 
        
    chemin_vers_sauvegarde_model = save_name 
    model = model.to(device) 
    
    
    loss_train = []
    loss_test = []
    train_accs = []
    test_accs = []
    
    
    for epoch in tqdm(range(n_epochs)):
        loss_train_epoch = 0
        train_acc = 0
        model.train()
        for batch_x,batch_y in dataloader_train:
    
            batch_x.to(device)
            batch_y.to(device)
            batch_y = batch_y.long()
    
            optimizer.zero_grad()
    
            batch_y_predicted = model(batch_x)
            #batch_y_predicted = batch_y_predicted.view(-1, batch_y_predicted.size(-1))  # [batch_size * seq_length, num_classes]
            
            batch_y = batch_y.view(-1)  # [batch_size * nb_outputs]
            #print(batch_y_predicted.shape, batch_y.shape)
            l = loss(batch_y_predicted, batch_y)
            loss_train_epoch += l
            train_acc += torch.sum(torch.argmax(batch_y_predicted, axis=1) == batch_y) / dataloader_train.batch_size
            # loggez la loss sur le batch d'entraînement
    
            l.backward()
    
            optimizer.step()
        train_accs.append(train_acc.detach().cpu().numpy() / len(dataloader_train))
        loss_train.append(loss_train_epoch.detach().cpu().numpy() / len(dataloader_train))
        model.eval()
        loss_test_epoch = 0
        test_acc = 0
        for batch_x,batch_y in dataloader_valid:
    
            batch_x.to(device)
            batch_y.to(device)
            
            with torch.no_grad():
                batch_y_predicted = model(batch_x)
                batch_y_predicted = (batch_y_predicted.view(-1, batch_y_predicted.size(-1)))  # [batch_size * seq_length, num_classes]
            
                batch_y = batch_y.view(-1).long()  # [batch_size * seq_length]
                loss_test_epoch +=  loss(batch_y_predicted, batch_y)
                batch_y_predicted = torch.argmax(batch_y_predicted, axis=1)
                test_acc += torch.sum(batch_y_predicted == batch_y) / dataloader_valid.batch_size
            # loggez la loss et les métriques sur le batch de validation
        test_accs.append(test_acc.cpu().numpy() / len(dataloader_valid))
        loss_test.append(loss_test_epoch.cpu().numpy() / len(dataloader_valid))
    torch.save(model, chemin_vers_sauvegarde_model)
    return loss_train, loss_test, train_accs, test_accs
````

## Premier Modèle

Nous avons choisi de baser notre premier modèle sur des couches de convolution 1D. Notre réseau est constitué de 6 couches convolution successives de taille de kernel progressivement décroissante de 15 à 11 avec un nombre de filtres croissant de 4 à 60. Entre chaque couche, nous avons ajouté une couche de batch normalization et une couche de max pooling.

A la fin du réseau nous appliquons une moyenne sur la dimension temporelle du signal puis une fonction d'activation LogSoftmax (correspondant simplement au log du softmax)

La majorité de nos choix découlent des nombreuses expériences que nous avons effectuées et notamment ceux-ci : 
- Nous avons préféré moyenner sur la dimension temporelle plutôt que d'applatir le vecteur entrainant un fort gain en performance
- La cross-entropy loss a été préférée à la negative log-likelyhood
- Le grand nombre de couches de convolution qui a été bénéfique aux prédictions
- L'utilisation de pooling à la place de strides sur les couches de convolution

### Définition du modèle

````python


class ConvNet1Dv2(torch.nn.Module):
    def __init__(self, input_channels = 2, N_Classes = 6, dropout_prop = 0.5):
        super(ConvNet1Dv2, self).__init__()

        self.conv1 = torch.nn.Conv1d(in_channels=input_channels, out_channels=4, kernel_size=15, padding="same")
        self.batchnorm1 = torch.nn.BatchNorm1d(num_features=4)

        self.conv2 = torch.nn.Conv1d(in_channels=4, out_channels=8, kernel_size=13, padding="same")
        self.batchnorm2 = torch.nn.BatchNorm1d(num_features=8)

        self.conv3 = torch.nn.Conv1d(in_channels=8, out_channels=16, kernel_size=11, padding="same")
        self.batchnorm3 = torch.nn.BatchNorm1d(num_features=16)

        self.conv4 = torch.nn.Conv1d(in_channels=16, out_channels=24, kernel_size=11, padding="same")
        self.batchnorm4 = torch.nn.BatchNorm1d(num_features=24)

        self.conv5 = torch.nn.Conv1d(in_channels=24, out_channels=40, kernel_size=11, padding="same")
        self.batchnorm5 = torch.nn.BatchNorm1d(num_features=40)

        self.conv6 = torch.nn.Conv1d(in_channels=40, out_channels=60, kernel_size=11, padding="same")
        self.batchnorm6 = torch.nn.BatchNorm1d(num_features=60)

        self.pool = torch.nn.MaxPool1d(kernel_size=2)  
        self.linear = torch.nn.Linear(in_features=60, out_features=N_Classes)
        self.relu = torch.nn.ReLU()
        self.softmax = torch.nn.LogSoftmax(dim=1)

    def forward(self, x):
        x = x.transpose(1, 2)
        # On applique les différentes couches de convolution-batchnorm-relu-maxpooling successives
        x = self.relu(self.batchnorm1(self.conv1(x)))
        x = self.pool(x)  

        x = self.relu(self.batchnorm2(self.conv2(x)))
        x = self.pool(x)  

        x = self.relu(self.batchnorm3(self.conv3(x)))
        x = self.pool(x) 

        x = self.relu(self.batchnorm4(self.conv4(x)))
        x = self.pool(x)  

        x = self.relu(self.conv5(x))
        x = self.pool(x) 

        x = self.relu(self.conv6(x))

        # Applique l'équivalent d'un Global Average Pooling sur la dimension temporelle
        x = torch.mean(x, axis=2)
        output = self.softmax(self.linear(x))
        return output
````

## Second modèle : mise en place d'un RNN

Comme nous travaillons sur des données séquentielles, un réseau récurrent pourrait être adapté au données séquentielles.

Le modèle prendra comme dimension d'entrée 2 (qui correspond au nombre de features).

Quant à 2048, c'est la longueur d'une séquence sur lequel le RNN va opérer. Donc la dimension attendue des données est nb_exemples x 2048 x 2.

Nous n'avons pas activé la bidirectionnalté car elle augmente drastiquement le nombre de paramètres.

L'utilisation d'un modèle de type GRU permet d'avoir un gradient plus stable (et donc un meilleur apprentissage).

### Démarrage en douceur : un modèle RNN de référence.

````python

class RNN_signal(nn.Module):
    def __init__(self, input_size = 2, hidden_size = 96, num_layers = 2, output_size = 6, sequence_length = 2048, dropout = 0.0):
        super(RNN_signal, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.rnn1 = nn.GRU(input_size, hidden_size, num_layers, batch_first = True, dropout = dropout, bidirectional = False)# batch_first = batch_size x seq_length x nb_features
        self.fc1 = nn.Linear(hidden_size , output_size)
        self.softmax = nn.LogSoftmax(dim = 1)
        #self.relu = torch.nn.ReLU()
        
    def forward(self, x, hidden_states = None):
        if hidden_states is None:
            hidden_states = torch.zeros(self.num_layers, x.size(0), self.hidden_size  ).to(device) # nb_layers x batch_size x hidden_size
        out, _ = self.rnn1(x, hidden_states)
        out = out[:, -1, :] # batch_size x sequence_length x hidden_size
        out = self.fc1(out)
        return self.softmax(out)
````

## Réseau à base de couches convolutionnelles et de couches récurrentes.

````python
import torch.nn.functional as F

class CNN_GRU_Classifier(nn.Module):
    def __init__(self, input_channels=2, cnn_filters=32, gru_hidden_size=64, gru_layers=2, num_classes=6, dropout=0.3):
        super(CNN_GRU_Classifier, self).__init__()
        
        # Étape CNN
        self.cnn1 = nn.Sequential(
            nn.Conv1d(input_channels, cnn_filters, kernel_size=10, stride=1, padding=2),
            nn.BatchNorm1d(num_features=cnn_filters),
            nn.ReLU(),
            nn.MaxPool1d(kernel_size=2, stride=2), 
            nn.Dropout(dropout)
        )
        self.cnn2 = nn.Sequential(
            nn.Conv1d(cnn_filters, 64, kernel_size=12, stride=1, padding="same"),
            nn.BatchNorm1d(num_features=64),
            nn.ReLU(),
            nn.MaxPool1d(kernel_size=2, stride=1), 
            nn.Dropout(dropout)
        )

        self.gru = nn.GRU(
            input_size=64,  
            hidden_size=gru_hidden_size,
            num_layers=gru_layers,
            batch_first=True,
            bidirectional=False,
            dropout=dropout
        )
        
        self.fc = nn.Linear(gru_hidden_size , num_classes)

    def forward(self, x):
        # x => (batch_size, seq_length, input_channels)
        #reformatage pour Conv1d: (batch_size, input_channels, seq_length)
        x = x.permute(0, 2, 1) 
        
        x = self.cnn1(x)  # (batch_size, cnn_filters, new_seq_length)
        x = self.cnn2(x)
        #reformatage pour GRU: (batch_size, new_seq_length, cnn_filters)
        x = x.permute(0, 2, 1)
        
        out, _ = self.gru(x)  # (batch_size, new_seq_length, gru_hidden_size*2)
        
        out = out[:, -1, :]  # (batch_size, gru_hidden_size*2)
        
        out = self.fc(out)  # (batch_size, num_classes)
        return F.log_softmax(out, dim=1)
````
