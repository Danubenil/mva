# Unsupervised Domain Adaptation

The three data-sets have the same features $\mathcal{X}$ (all images have size $[32 x 32]$)
 and the same classes (output labels) $\mathcal{Y}$ but they have different marginal $p(X)$ distributions.

In this part, we will assume that we only have labeled data in the source domain $S$ but not int the target domain $T$. By matching the source $p_S(X)$ and target $p_T(X)$ marginal distributions, one can hope that a classifier learnt on $S$ will also work on $T$.

Here, you will implement and test the following algorithm based on subspace alignement:

B. Fernando et al. "*Unsupervised Visual Domain Adaptation Using Subspace Alignment*". ICCV, 2013.

The first thing to do is to choose the source domain and the target domain.
Let's start with the MNIST dataset as source and USPS as target.

We will use the previousyly computed reduced datasets to speed up computations.

```python
# Source domain
XtrainS = MNIST_Xtrain_res
ytrainS = MNIST_ytrain_res
XtestS = MNIST_Xtest
ytestS = MNIST_ytest

# # Target domain
XtrainT = USPS_Xtrain_res
ytrainT = USPS_ytrain_res
XtestT = USPS_Xtest
ytestT = USPS_ytest

# Target domain
# XtrainT = SVHN_Xtrain_res
# ytrainT = SVHN_ytrain_res
# XtestT = SVHN_Xtest
# ytestT = SVHN_ytest

Ns=1000 # number of samples in source domain
Nt=1000 # number of samples in target domain
```

Implement the method following the description of the method in the slides of the lecture (around slide 75).  

```python
from sklearn.decomposition import PCA

def SA(XTrainS,XTestS,XTrainT,XTestT,d=10):
    """
    Subspace Alignment (SA)
    
    Input
    XTrainS : Input training source data [NTrS,D]
    XTestS : Input test source data [NTeS,D]
    XTrainT : Input training target data [NTrT,D]
    XTestT : Input test target data [NTeT,D]
    d : number of PCA components
    
    Output
    XTrainSp : projected and transformed source training data [NTrS,d]
    XTestSp : projected and transformed source test data [NTeS,d]
    XTrainTp : projected training target data [NTrT,d]
    XTestTp : projected test target data [NTeT,d]
    """
    pca = PCA(d)
    pca.fit(XTrainS)
    Us_train = pca.components_.T
    pca.fit(XTestS)
    Us_test = pca.components_.T
    pca.fit(XTrainT)
    Ut_train = pca.components_.T
    pca.fit(XTestT)
    Ut_test = pca.components_.T
    M_train = Us_train.T @ Ut_train
    M_test = Us_test.T @ Ut_test
    XTrainSp = XTrainS @ Us_train @ M_train
    XTrainTp = XTrainT @ Ut_train
    XTestSp = XTestS @ Us_test @ M_test
    XTestTp = XTestT @ Ut_test 
    return XTrainSp,XTestSp,XTrainTp,XTestTp
```

```python
XTrainSp,XTestSp,XTrainTp,XTestTp = SA(XtrainS, XtestS, XtrainT, XtestT)
```

We start by veryfying the score of a classifier trained on the source domain and tested on the target domain (no adaptation).

You can use the classifier you want (e.g., logistic regression, SVM, ...)

```python
model = SVC()
model.fit(XTrainSp, ytrainS)
print(f"Score on source trained on source : {model.score(XTestSp, ytestS):.3f}")
print(f"Score on target trained on source : {model.score(XTestTp, ytestT):.3f}")
model.fit(XTrainTp, ytrainT)
print(f"Score on target trained on target : {model.score(XTestTp, ytestT):.3f}")
```

Results are really bad, this is due to the fact that we're training PCA separately on the train and on the test, which doesn't help for generalization.

To reduce dimensionality, we can also first compute PCA using the source training set and then project (i.e., transform) all data (source and target, training and test) on the first $d$ components.

**Question**: try different values of $d$. Is there a difference with the previous full-training data appraoch ?

```python
from sklearn.decomposition import PCA

def SA(XTrainS,XTestS,XTrainT,XTestT,d=10):
    """
    Subspace Alignment (SA)
    
    Input
    XTrainS : Input training source data [NTrS,D]
    XTestS : Input test source data [NTeS,D]
    XTrainT : Input training target data [NTrT,D]
    XTestT : Input test target data [NTeT,D]
    d : number of PCA components
    
    Output
    XTrainSp : projected and transformed source training data [NTrS,d]
    XTestSp : projected and transformed source test data [NTeS,d]
    XTrainTp : projected training target data [NTrT,d]
    XTestTp : projected test target data [NTeT,d]
    """
    pca = PCA(d)
    pca.fit(XTrainS)
    Us_train = pca.components_.T
    Us_test = pca.components_.T
    pca.fit(XTrainT)
    Ut_train = pca.components_.T
    Ut_test = pca.components_.T
    M_train = Us_train.T @ Ut_train
    M_test = Us_test.T @ Ut_test
    XTrainSp = XTrainS @ Us_train @ M_train
    XTrainTp = XTrainT @ Ut_train
    XTestSp = XTestS @ Us_test @ M_test
    XTestTp = XTestT @ Ut_test 
    return XTrainSp,XTestSp,XTrainTp,XTestTp
```

We've lost 3 % accuracy on source domain, and 13 % on the target domain

```python
factors = [1.5, 2, 2.5, 3, 3.5, 4, 6,  8, ]
mlp = MLPClassifier((100, 50),alpha=0.001, max_iter=1000)
d = 100
scores_source_mlp = []
scores_target_mlp = []
for factor in tqdm(factors):
    # Source domain
    indice = int(len(MNIST_Xtrain_res) / factor)
    XtrainS = MNIST_Xtrain_res[:indice]
    ytrainS = MNIST_ytrain_res[:indice]
    XtestS = MNIST_Xtest
    ytestS = MNIST_ytest
    
    # # Target domain
    XtrainT = USPS_Xtrain_res[:indice]
    ytrainT = USPS_ytrain_res[:indice]
    XtestT = USPS_Xtest
    ytestT = USPS_ytest
    XTrainSp, XTestSp, XTrainTp, XTestTp = SA(XtrainS, XtestS, XtrainT, XtestT, d)
    mlp.fit(XTrainSp, ytrainS)
    scores_source_mlp.append(mlp.score(XTestSp, ytestS))
    scores_target_mlp.append(mlp.score(XTestTp, ytestT))
```

```python
plt.xlabel("Divisive factor")
plt.ylabel("Accuracy score")
plt.title("Accuracy score as a function of the amount we divided the dataset length by")
plt.plot(factors, scores_source_mlp, label = "on_source")
plt.plot(factors, scores_target_mlp, label = "on_target")
```

The two accuracy scores are decreasing, even though the accuracy score for the source domain is decreasing faster and in a steady manner.

This is related to paper "J. Blitzer et al. “Learning Bounds for Domain Adaptation”", where the complexity of the model increases with the decrease of the number of source and target training samples.

Thus, we are slowly underfitting our data.
