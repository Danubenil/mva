```python
frequencies = np.array([[1209, 1336, 1477, 1633], [697, 770, 852, 941]]).flatten()
dtmf = {
    '1': (697, 1209),
    '2': (697, 1336),
    '3': (697, 1477),
    'A': (697, 1633),
    '4': (770, 1209),
    '5': (770, 1336),
    '6': (770, 1477),
    'B': (770, 1633),
    '7': (852, 1209),
    '8': (852, 1336),
    '9': (852, 1477),
    'C': (852, 1633),
    '*': (941, 1209),
    '0': (941, 1336),
    '#': (941, 1477),
    'D': (941, 1633)
}
dtmf_inv = {v : k for k,v in dtmf.items()}
```

```python
from scipy.signal import spectrogram

def get_sequence(signal, nperseg = 1024, jump = 1, model = "rbf", min_size = 3, display = True, eps = 0.03):
  """
  Get the sequence of symbols that signal encodes.
  --Args:
    signal
    nperseg (Length of each segment for the spectrogram )
    pen
    jump
    model
    min_size
    display
    eps : threshold to keep a window of the changepoint method
  --Returns:
    classes
    tranches
  """

  # SFT
  f, t, Zxx  = spectrogram(signal, fs = FS, nperseg = nperseg)
  min_f = 500
  max_f = 1750
  indice_freq_min = np.argmin(np.abs(f - min_f ))
  indice_freq_max = np.argmin(np.abs(f - max_f ))
  # On récupère les deux fréquences dans la sft les plus proches de chaque fréquence de la DTMF
  nearest = np.abs(f - frequencies.reshape(-1, 1))
  nearest = nearest.argsort(axis = 1 )
  indexs = nearest[:, :2]
  # On supprime les valeurs en dehors des fréquences intéressantes.
  mask = np.ones_like(f, dtype=bool)
  mask[indexs] = False
  Zxx[mask] = 0


  whole_signal = (np.sum(Zxx, axis = 0))
  sigma = np.std(whole_signal)
  n = len(whole_signal)
  algo =  rpt.Binseg(model=model, min_size = min_size, jump = jump).fit( whole_signal)
  result = algo.predict(pen= 3 * n * sigma ** 2)
  if result[0] != 0:
    result.insert(0, 0)
  result = np.array(result)
  if display:
    plt.pcolormesh(t, f, np.abs(Zxx), shading='gouraud')
    plt.title('STFT Magnitude')
    plt.ylabel('Frequency [Hz]')
    plt.xlabel('Time [sec]')
    plt.show()
    rpt.display( whole_signal, [0], result)
    plt.show()
  aggs = []
  res_tuple = []
  for i in range(len(result) - 1):
    agg = whole_signal[result[i]:result[i+1]].max()
    aggs.append(agg)
    if agg > eps:
      res_tuple.append((result[i], result[i+1]))

  classes = []
  tranches = []
  for i in range(len(res_tuple)):

    start, end = res_tuple[i]
    portion = np.mean(Zxx[:, start:end], axis = 1) # On somme sur les fréquences

    argportion = np.argsort(portion)
    freqs_detectees = f[argportion[-2:]] # Les deux plus grandes fréquences
    # On cherche les fréquences correspondantes

    diff = np.abs(frequencies.reshape(-1, 1)- freqs_detectees )
    i_f1 = diff[:, 0].argsort()
    i_f2 = diff[:, 1].argsort()

    for k in i_f1:
      for l in i_f2:

        correspondances = frequencies[[k, l]]
        correspondances.sort()
        correspondances = tuple(correspondances)
        if correspondances in dtmf_inv:
          break
      if correspondances in dtmf_inv:
        break
    classes.append(dtmf_inv[correspondances])
    if end == len(t):
      end = end - 1
    tranches.append((t[start], t[end]))
  return classes, tranches
```

```python
r , new_t = get_sequence(X_train[4])
```

```python
from tqdm import tqdm
def accuracy(X, Y):
  dist = 0
  for x, y in tqdm(zip(X, Y)):
    seq, t = get_sequence(x, display = False)
    dist += 1 - Levenshtein.distance(seq, y) / (max(len(seq), len(y)))
  dist /= len(X)
  return dist
```

```python
M = 9
def gu(lambd, a, R):
  gu = 0.5 * (1 + np.cos(2 * np.pi * (lambd / (a * R) + 0.5))) * (lambd < 0) * (-R* a  <= lambd)
  return gu
def gm(lambd, lambd_n, m, R):
  a = lambd_n / (M + 1 - R)
  return gu(lambd - a * m, a, R )
```

```python
g_R1 = []
g_R2 = []
g_R3 = []
lambd = np.linspace(-25, 20, 1000)
lambd_n = 12
for i in range(1, M):
  g_R1.append(gm(lambd, lambd_n, i, 1))
  g_R2.append(gm(lambd, lambd_n, i, 3))
  g_R3.append(gm(lambd, lambd_n, i, 5))
g_R1 = np.array(g_R1)
g_R2 = np.array(g_R2)
g_R3 = np.array(g_R3)
g_R = [g_R1, g_R2, g_R3]
```

```python
fig, axs = plt.subplots(1, 3, figsize=(58, 10))
for i, R in enumerate([1, 3, 5]):
  for m in range(0, M - 1):
    axs[i].plot(lambd, g_R[i][m], label = f"m = {m + 1}")
  axs[i].set_xlabel("lambda")
  axs[i].set_ylabel("g_m(lambda)")
  axs[i].set_title(f"R = {R}")
  axs[i].legend()
  portion = axs[i].get_window_extent().transformed(fig.dpi_scale_trans.inverted())
  fig.savefig(f'R = {R}', bbox_inches=portion)
```

```python
def get_geodesic_distance(point_1, point_2) -> float:
    """
    Calculate the great circle distance (in km) between two points
    on the earth (specified in decimal degrees)

    https://stackoverflow.com/a/4913653
    """

    lon1, lat1 = point_1
    lon2, lat2 = point_2

    # convert decimal degrees to radians
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(dlon / 2) ** 2
    c = 2 * asin(sqrt(a))
    r = 6371  # Radius of earth in kilometers. Use 3956 for miles
    return c * r
def get_exponential_similarity(condensed_distance_matrix, bandwidth, threshold):
  exp_similarity = np.exp(-(condensed_distance_matrix**2) / bandwidth / bandwidth)
  res_arr = np.where(exp_similarity > threshold, exp_similarity, 0.0)
  return res_arr
```

```python
def W(signal, eigenvals, eigenvect):
  n = len(signal)
  argind = eigenvals.argsort()
  eigenvals = eigenvals[argind]
  eigenvect = eigenvect[:, argind]
  lambda_n = eigenvals[-1]
  W_res = np.zeros((len(signal), M) )
  fourier = eigenvect.T @ signal
  for v in range(n):
    for m in range(1, M + 1):
      g = gm(eigenvals, eigenvals[-1] , m, R)
      W_res[v, m - 1] = np.sum(g * fourier * eigenvect[:, v], axis = 0)
  return W_res
```

```python
def plot_classifications_on_map(graph, stations_gdf, stations_np, classifications, title):
    """
    Plot node classifications on a map.

    Args:
    - graph: PyGSP Graph object for the graph structure.
    - stations_gdf: GeoDataFrame of stations, used to plot the geographic map.
    - stations_np: Node positions (array of shape [num_nodes, 2]).
    - classifications: List of classifications for each node (e.g., "low", "medium", "high").
    - title: Title for the plot.
    """
    class_map = {"low": 1, "medium": 2, "high": 3}
    signal = [class_map[classification] for classification in classifications]
    signal = np.array(signal)

    ax = stations_gdf.geometry.plot(figsize=(10, 6))
    cx.add_basemap(ax, crs=stations_gdf.crs.to_string(), zoom=8)
    ax.set_axis_off()

    graph.set_coordinates(stations_np)
    graph.plot_signal(signal, ax=ax)

    ax.set_title(title, fontsize=15)
    print("Nodes classified as low (1) are blue, medium (2) are green, and high (3) are yellow.")
    plt.savefig(title)
    plt.show()

plot_classifications_on_map(G, stations_gdf, stations_np, classes_str[2], "First Timestamp Signal Classification")
plot_classifications_on_map(G, stations_gdf, stations_np, classes_str[0], "Smoothest Signal Classification")
plot_classifications_on_map(G, stations_gdf, stations_np, classes_str[1], "Least Smooth Signal Classification")
```

```python
#Compute eigenvalues and eigenvectors of G and G_prime
eigenvalues_G, eigenvectors_G = eigh(Laplacian_G)
eigenvalues_G_prime, eigenvectors_G_prime = eigh(Laplacian_G_prime)

#Compute all eignevalues of H (all sums between eigenvalues of G and G_prime)
eigenvalues_H, eigenvectors_H = [],[]
for eigenvalue_G, eigenvector_G in zip(eigenvalues_G,eigenvectors_G):
    for eigenvalue_G_prime, eigenvector_G_prime in zip(eigenvalues_G_prime,eigenvectors_G_prime):
        eigenvalues_H.append(eigenvalue_G+eigenvalue_G_prime)
        eigenvectors_H.append(np.kron(eigenvector_G,eigenvector_G_prime))

```
